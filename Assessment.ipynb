```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://www.tnprivatejobs.tn.gov.in/Home/jobs"

all_jobs = []
num_pages = 2

response = requests.get(base_url)
if response.status_code ==200:

    for page in range(1, num_pages+1):

        if page == 1:
            url = base_url
        else:
            url = f'{base_url}/{(page-1)*10}'
        
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        job_listings = soup.find(class_='searchList')

        print
        job_details = job_listings.find_all('li')

        for job in job_details:
            job_title = job.find('h4').get_text(strip=True)
            company_name_div = job.find('div', class_='companyName')
            company_name = company_name_div.find('a').get_text(strip=True)
            job_role = company_name_div.contents[2].strip().rstrip('|').strip()

            location_div = job.find('div', class_='location')
            location_salary = location_div.get_text(strip=True).split("|") 
            salary = location_salary[0].strip()
            location = location_salary[1].strip()
            sector = location_salary[2].strip()







            all_jobs.append({
                "Job Title": job_title,
                "Company Name": company_name,
                "Job Role": job_role,
                # "Posted Date": posted_date,
                # "Open Until": open_until,
                "Salary": salary,
                "Location": location,
                "sector": sector
            })







df = pd.DataFrame(all_jobs)

# Write the DataFrame to a CSV file
df.to_csv('tn_private_jobs.csv', index=False)

print("Data has been written to tn_private_jobs.csv")  

     







# job_list = soup.find(class_='searchList')
# # print(job_list.prettify())

# job_name = job_list.find_all('h4')

# for name in job_name:
#     print(name.get_text().strip()) 

```

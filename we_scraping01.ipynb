```python
import requests
from bs4 import BeautifulSoup
```

```python
def fetchAndSave(url, path):
    r  = requests.get(url)
    with open(path,"w",encoding="utf-8") as f:
        f.write(r.text)

```

```python
url = "https://timesofindia.indiatimes.com/city/delhi"
```

```python
fetchAndSave(url, "times.html")
```
